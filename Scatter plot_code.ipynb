```python
import dash
import dash_core_components as dcc
import plotly.graph_objs as go
import dash_html_components as html
import pandas as pd
import numpy as np
```

```python
app = dash.Dash()
```

```python
orders = pd.read_csv('sales.csv')
```

```python
orders
```

```python
orders.columns.tolist( )
```

```python
app.layout = html.Div([
            
    
    dcc.Graph(
        id = 'scatter_chart',
        figure = {
            'data' : [
                go.Scatter(
                    x = orders['Country'],
                    y = orders['Total Profit'],
                    mode = 'markers'
                    
                    )
                
            ],
            'layout' : go.Layout(
                title = 'Scatterplot of random 60 points',
                xaxis = { 'title' : 'Sales'},
                yaxis = { 'title' : 'Profit'}
            
            )
        }
        
        

        
        
    )
])
```

```python
if __name__ == '__main__':
    app.run_server(port=4050)
```
